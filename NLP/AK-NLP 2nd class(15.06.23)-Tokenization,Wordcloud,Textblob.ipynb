```python
import nltk
nltk.download()
```

```python
editorial="In the aftermath of a conflict, a return to normalcy must be a necessary basis to address the grievances of the affected parties that led to the conflict in the first place. There cannot be any redress when groups engage in wanton and targeted violence against each other, use blockades to prevent the transport of supplies, and continue to prevent displaced people from getting back to their homes. Return to normalcy is the first step before grievances are heard to prevent another conflict from emerging. And in order to effect the first step, peace talks between eminent representatives — from civil society and political parties — in the conflict and tamping down any embers of violence that might continue following the conflagration are a must. The presence of paramilitary forces and the appeal for peace by Union Home Minister Amit Shah after a visit to the tribal and the Imphal valley areas have helped stem the violence to an extent even if sporadic incidents of ethnic violence and arson have continued a month after the conflagration in Churachandpur, Imphal and other areas. But in terms of setting the conditions of peace which would entail the return of looted arms and the slow and sure return of those displaced to their battered homes, there has been little progress."
#write the paragraph in a single line
```

```python
editorial
```

# Tokenization

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
#sentence tokenization

sent_tokenize(editorial)
```

```python
for sent in sent_tokenize(editorial):
    print(sent)
```

```python
# Word Tokenization

word_tokenize(editorial)
```

```python
for word in word_tokenize(editorial):
    print(word)
```

# Whitespace Tokenizer

```python
white_space = nltk.tokenize.WhitespaceTokenizer()
```

```python
white_space.tokenize(editorial)
```

```python
for white in white_space.tokenize(editorial):
    print(white)
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_stop =stopwords.words('english')
word_stop
```

```python
wcloud=WordCloud(max_words=100,stopwords=word_stop)
wcloud.generate(complete_text)
```
