```python
import nltk
import pandas as pd
editorial="Other higher education institutions should follow IIT Bombay’s example, make aggressive fundraising efforts:Nandan Nilekani’s $38. 5 million donation to his alma mater IIT Bombay says many good things about him and will get many good things done for the institution.But it raises the question why many of India’s public education institutes are so lagging in fundraising. This is a particularly sharp question since there’s even a designated body to help raise funds. HEFA, founded in 2017, is a non-profit lending company for higher education institutes’ infrastructure projects. By the end of last year, HEFA had approved 141 projects worth just `46,700 crore, against the target of `1 lakh crore. HEFA would raise funds from private sector – that was the plan, but no details are available. GOI as equity contributor put in `1 lakh in 2022-23, and zero in 2023-2024. That’s how slow the take-off is.Funds crunch is hobbling higher education – 11,000 teaching posts are vacant across institutions, including central universities. When fees are jacked up, students either drop out or move overseas, as in the case of medical students. Since 2017, over 30 lakh students have gone abroad, 7. 5 lakh in 2021 alone. GOI last year asked central universities to set up endowments, including FCRA accounts. Even the national capital’s Delhi University is struggling to get funds. The reality is that if institutions want to grow and want to become world-class, reliance on only government funding will get them nowhere.The source to tap is private wealth. India has 169 billionaires; 1,103 individuals with `1,000 crore-plus wealth. NRI corporate leaders and tech czars are more than generous in giving endowments to American universities. But most of them have contributed little to India’s institutions. One reason for that is that the latter don’t have effective sales pitches. IIT Bombay is holding roadshows. Others should follow."
```

```python
editorial
```

```python
import re  #re=regular expresssion
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

lemm=WordNetLemmatizer()

sentence=nltk.sent_tokenize(editorial)
```

```python
sentence
```

```python
# Clean your text before applying feature engineering/extraction method

corpus=[]

for i in range(len(sentence)):
    words=re.sub('[^a-zA-Z]',' ', sentence[i])
    words=words.lower()
    words=words.split()
    words=[lemm.lemmatize(word) for word in words if not word in set(stopwords.words('english'))]
    words=' '.join(words)
    corpus.append(words)
```

```python
corpus
```

# Build TF-IDF Feature Extraction

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(corpus).toarray()
```

```python
tfidf.vocabulary_
```

```python
pd.DataFrame(x)
```
