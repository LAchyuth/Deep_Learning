# NLTK - Lemmatization

```python
from nltk.stem import WordNetLemmatizer

# creating the object for lemmatizer
lemmatizer = WordNetLemmatizer()

print("dancing :", lemmatizer.lemmatize("dancing", pos="a"))
```

```python
words = ['happy','happier','happiest','happyness','run','runs','ran','running']
```

```python
for word in words:
    print(word + '----->' + lemmatizer.lemmatize(word))
```

```python
for word in words:
    print(word + '----->' + lemmatizer.lemmatize(word, pos="a"))
```

# NLTK - Stemming

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

print("helping:", ps.stem("helping"))
```

```python
print("caring:", ps.stem("caring"))
```

```python
for word in words:
    print(word + '----->' + ps.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
```

```python
for word in words:
    print(word + '----->' + s_stemmer.stem(word))
```
